```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
df = pd.DataFrame(school_data)
school_name = df["school_name"]
unique_school_count = len(school_name)
unique_school_count
```

```python
# Calculate the total number of students
df = pd.DataFrame(student_data)
student_name = df["student_name"]
student_count = len(student_name)
student_count
```

```python
# Calculate the total budget
budget = school_data["budget"]
total_budget = budget.sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Calculate the percentage of students who passed math
passing_threshold = 70
total_students = len(school_data_complete)
passed_math_students = len(school_data_complete[school_data_complete["math_score"] >= passing_threshold])
passing_math_percentage = (passed_math_students / total_students) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading
passing_threshold = 70
total_students = len(school_data_complete)
passed_reading_students = len(school_data_complete[school_data_complete["reading_score"] >= passing_threshold])
passing_reading_percentage = (passed_reading_students / total_students) * 100
passing_reading_percentage
```

```python
 #Calculate the percentage of students that passed math and reading
total_students = len(school_data_complete)
passed_math_and_reading_students = len(
    school_data_complete[(school_data_complete["math_score"] >=70) & 
                         (school_data_complete["reading_score"] >=70)])
percentage_passed_math_and_reading = (passed_math_and_reading_students / total_students) * 100
percentage_passed_math_and_reading
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [unique_school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Passing": [percentage_passed_math_and_reading]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```
